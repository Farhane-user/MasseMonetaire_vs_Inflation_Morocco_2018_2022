```python
import pandas as pd

df_1 = pd.read_csv(r"C:\Users\jfarh\OneDrive\Desktop\Money_Supply_M1_M2_M3_Morocco.csv")
df_2 = pd.read_excel(r"C:\Users\jfarh\OneDrive\Desktop\Inflation_et_inflation_sous_jacente.xlsx")
```

```python
df_1['date'] = pd.to_datetime(df_1['date'])
df_1['date'] = df_1['date'].dt.strftime('%b/%Y')

df_2['date'] = pd.to_datetime(df_2['date'])
df_2['date'] = df_2['date'].dt.strftime('%b/%Y')  
```

```python
df_1c= df_1.tail(60)
```

```python
merged_df = pd.merge(df_1c, df_2, on='date', how='inner')
merged_df.rename({'date':'Date'},inplace=True)
```

```python
merged_df['M1_diff'] = merged_df['M1'].diff()
merged_df['M1_change'] = merged_df['M1'].pct_change() * 100

merged_df['M2_diff'] = merged_df['M2'].diff()
merged_df['M2_change'] = merged_df['M2'].pct_change() * 100

merged_df['M3_diff'] = merged_df['M3'].diff()
merged_df['M3_change'] = merged_df['M3'].pct_change() * 100
```

```python
merged_df['High_Inflation'] = merged_df['Inflation'].apply(lambda x: 'Yes' if x > 5 else 'No')
merged_df.head()
```
